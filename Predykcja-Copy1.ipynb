```python
def MacierzSpokrewnien(s, d):
    if len(s) != len(d):
        stop("Length of sire oraz dtam vectors is different")
    else:
        n = len(s)
        N = n + 1
        A = np.zeros((N,N))
        s = (s == 0.0)* (N) + s
        d = (d == 0.0)* N + d
        for i in range(n):
            print(s[i],d[i],A[s[i]][d[i]])
            A[i][i] = 1.0 + A[s[i]][d[i]]/2.0
            for j in (i+1,n):
                if (j >= n):
                    break
                A[i][j] = (A[i][s[j]] + A[i][d[j]])/2.0
                A[j][i] = A[i][j] 
              
        return(A)
       
```
